```python
import pandas as pd
import numpy as np
```

# Creating dataframes

```python
sunscreen = pd.read_csv('data/sunscreen.csv')
sunscreen['product'] = 'sunscreen'
sunscreen.head()
```

```python
cleanser = pd.read_csv('data/cleanser.csv')
cleanser['product'] = 'cleanser'
cleanser.head()

# data cleaning
cleanser['price'] = cleanser['price'].str.replace('$', '').astype(float)
cleanser['ys_num_reviews'] = cleanser['ys_num_reviews'].fillna(0).astype(int)
cleanser['az_num_reviews'] = cleanser['az_num_reviews'].fillna(0).astype(int)
cleanser.head()
```

```python
cleanser.loc[0, 'image']
```

```python
essence = pd.read_csv('data/essence.csv')
essence['product'] = 'essence'
essence.head()
```

```python
moisturizer = pd.read_csv('data/moisturizer.csv')
moisturizer['product'] = 'moisturizer'
moisturizer.head()
```

```python
toner = pd.read_csv('data/toner.csv')
toner['product'] = 'toner'
toner.head()
```

## Dataframe of all products

```python
products = pd.concat([sunscreen, cleanser, essence, moisturizer, toner], ignore_index=True, axis = 0)
products
```
